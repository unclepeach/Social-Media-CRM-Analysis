```python
import os
import pandas as pd
```

```python
os. chdir("/Users/UNclePeach/Downloads/mailchimp")
path = os.getcwd()
files = os.listdir(path)
#files
```

```python
files_xlsx = [f for f in files if f[-4:] == 'xlsx']
#files_xlsx
```

```python
df = pd.DataFrame()
for f in files_xlsx:
    data = pd.read_excel(f, 'inventory')
    df = df.append(data)
    df = df.drop(columns = ['Organization'])
#df
```
